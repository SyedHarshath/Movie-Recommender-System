```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
# Features to be selected
# genres
# id
# keywords
# title (original title may have other language's title - not unique)
# overview
# cast
# crew
```

```python
movies['original_language'].value_counts()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
# Helper function to combine all the genres in the dictionary into one
def convert(objects):
    import ast
    genres = []
    for obj in ast.literal_eval(objects):
        genres.append(obj['name'])
    return genres
```

```python
print(convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'))
```

```python
# To convert this string of lists '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# into list use 'ast'
```

```python
import ast
print(ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'))
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head(1)
```

```python
def convert_cast(objects):
    genres = []
    counter = 0
    for obj in ast.literal_eval(objects):
        if counter != 3:
            genres.append(obj['name'])
            counter += 1
        else:
            break
    return genres
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head(1)
```

```python
def fetch_director(objects):
    director = []
    for obj in ast.literal_eval(objects):
        if obj['job'] == 'Director':
            director.append(obj['name'])
            break
    return director
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(1)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_movies = movies[['movie_id','title','tags']]
```

```python
new_movies.head(1)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies.head(1)
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x:x.lower())
```

```python
new_movies.head(1)
```

```python
new_movies['tags'][0]
```

```python
new_movies['tags'][1]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for word in text.split():
        y.append(ps.stem(word))
    return " ".join(y)
```

```python
new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_movies['tags'])
```

```python
import pickle
pickle.dump(vectors,open("vectors.pkl","wb"))
```

```python
vectors
```

```python
# In higher dimensional data, using Euclidean Distance is not a reliable measure
# So here I use cosine distance or similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity
```

```python
similarity[0]
# Similarity of first movie to all the other 4806 movies
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for movie in movies_list:
        print(new_movies.iloc[movie[0]].title)
```

```python
recommend("Pirates of the Caribbean: At World's End")
```
